```python
import geopandas as gpd
import matplotlib.pyplot as plt
import shapely.geometry
import numpy as np
import geog
import pandas as pd
```

```python
shp_path_node = '/content/drive/MyDrive/NODELINKDATA/MOCT_NODE.shp'
shp_path_link = '/content/drive/MyDrive/NODELINKDATA/MOCT_LINK.shp'
sf_node = gpd.read_file(shp_path_node)
sf_node = sf_node.to_crs(epsg=4326)
```

```python
point = shapely.geometry.Point([126.99082244729578, 37.54886890032513])

# 반경 20km
num_point = 200
radius = 20000
angles = np.linspace(0, 360, num_point)
circle = shapely.geometry.Polygon(geog.propagate(point, angles, radius))
```

```python
arr = []
for sf in range(len(sf_node)):
    if sf_node['geometry'][sf].intersects(circle) == True:
        arr.append(sf)
df = pd.DataFrame(sf_node.iloc[i] for i in arr)
```

```python
for i in range(len(df)):
    geo = str(df.iloc[i]['geometry'])
    geo = geo.replace('(', '')
    geo = geo.replace(')', '')
    geo = geo.split()
    geo = geo[1] + " " + geo[2]
    df.iloc[i]['geometry'] = geo
df
```

```python
df = df.reset_index()
df = df.drop(['index'],axis=1)
df.to_csv('node.csv', index=False, encoding='utf-8-sig')
```

```python
sf_link = gpd.read_file(shp_path_link)
sf_link = sf_link.to_crs(epsg=4326)
```

```python
arr2 = []
for sf in range(len(sf_link)):
    if sf_link['geometry'][sf].intersects(circle) == True:
        arr2.append(sf)
df2 = pd.DataFrame(sf_link.iloc[i] for i in arr2)
```

```python
df2 = df2.reset_index()
df2 = df2.drop(['index'], axis=1)
df2.to_csv('link.csv', index=False, encoding='utf-8-sig')
```
